```python
import pandas as pd
import numpy as np
import math as math
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("https://data.wprdc.org/datastore/dump/0f214885-ff3e-44e1-9963-e9e9062a04d1")
data
```

```python
cases = data.drop(columns=['update_date','indv_tested'])
data1 = cases.sort_values('cases',ascending=True)
data1_1 = data1.head(110)
data1_2 = data1.tail(110)
```

```python
data1_1.plot(figsize=(20,10),kind='bar',stacked=True,x='neighborhood_municipality',title='COVID-19 Cases and Deaths Based on Neighborhood').legend(bbox_to_anchor=(1.5,1))
```

```python
data1_2.plot(figsize=(20,10),kind='bar',stacked=True,x='neighborhood_municipality',title='COVID-19 Cases and Deaths Based on Neighborhood').legend(bbox_to_anchor=(1.5,1))
```
