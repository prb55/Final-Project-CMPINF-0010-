```python
import pandas as pd
import numpy as np
```

```python
dtype={'police_zone': 'string'}
low_memory=False

pgh_trees_data = pd.read_csv("https://data.wprdc.org/datastore/dump/1515a93c-73e3-4425-9b35-1cd11b2196da",index_col="id")
```

```python
# Creating a query mask that consists of only the police zone 
query_mask = pgh_trees_data[['police_zone']]

# Loops through the dataset and counts up every instance of a tree in each zone, prints it out
zone_1 = 0
for row in range (0, 45709):
    temp = query_mask.iat[row, 0]
    if (temp != 1):
        zone_1 = zone_1 + 1    
print("Total number of trees in zone 1:" , 45710 - zone_1)

zone_2 = 0
for row in range (0, 45709):
    temp = query_mask.iat[row, 0]
    if (temp != 2):
        zone_2 = zone_2 + 1    
print("Total number of trees in zone 2:" , 45710 - zone_2)

zone_3 = 0
for row in range (0, 45709):
    temp = query_mask.iat[row, 0]
    if (temp != 3):
        zone_3 = zone_3 + 1    
print("Total number of trees in zone 3:" , 45710 - zone_3)

zone_4 = 0
for row in range (0, 45709):
    temp = query_mask.iat[row, 0]
    if (temp != 4):
        zone_4 = zone_4 + 1    
print("Total number of trees in zone 4:" , 45710 - zone_4)

zone_5 = 0
for row in range (0, 45709):
    temp = query_mask.iat[row, 0]
    if (temp != 5):
        zone_5 = zone_5 + 1    
print("Total number of trees in zone 5:" , 45710 - zone_5)

zone_6 = 0
for row in range (0, 45709):
    temp = query_mask.iat[row, 0]
    if (temp != 6):
        zone_6 = zone_6 + 1    
print("Total number of trees in zone 6:" , 45710 - zone_6)


#zone 1 - brighton heights, manchesster, perry south
#zone 2 - polish hill, hill district, strip district, downtown
#zone 3 - south side, mt washington
#zone 4 - squirrel hill north and south
#zone 5 - east liberty and highland park
#zone 6 - crafton heights, elliott, duquesne heights
```
